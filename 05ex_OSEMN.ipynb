1\. Create a random list of number and then save it to a text file named "simple_data.txt"

```python
import random

# Generate a random list of numbers
random_numbers = [random.randint(1, 100) for _ in range(10)]

# Save the list to a text file
file_path = "simple_data.txt"
with open(file_path, "w") as file:
    for number in random_numbers:
        file.write(str(number) + "\n")

print(f"Random list of numbers saved to {file_path}")
```

2\. Create a random matrix of 5x5 and then save it to a text file named "data.txt"

```python
import random

# Generate a random 5x5 matrix
random_matrix = [[random.randint(1, 100) for _ in range(5)] for _ in range(5)]

# Save the matrix to a text file
file_path = "data.txt"
with open(file_path, "w") as file:
    for row in random_matrix:
        file.write("\t".join(map(str, row)) + "\n")

print(f"Random 5x5 matrix saved to {file_path}")
```

4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
# Specify the binary file name
binary_file_name = "credit_card.dat"

# Read the binary file
with open(binary_file_name, "rb") as binary_file:
    # Read the binary data
    binary_data = binary_file.read()

# Initialize an empty string to store the credit card numbers
credit_card_numbers = ""

# Loop through the binary data, extracting 6-bit chunks
for i in range(0, len(binary_data), 6):
    # Extract a 6-bit chunk
    six_bits = binary_data[i:i+6]

    # Convert the 6-bit chunk to a decimal number
    decimal_number = int(six_bits, 2)

    # Check if the decimal number is within the valid ASCII range
    if 32 <= decimal_number <= 126:
        # Convert the decimal number to a character and append to the result string
        credit_card_numbers += chr(decimal_number)

# Print the extracted credit card numbers
print("Extracted Credit Card Numbers:")
print(credit_card_numbers)
```

5\. Load the file "user_data.json", filter the data by the "CreditCardType" field equals to "American Express". Than save the data a to CSV.

```python
import json
import csv

# Load the JSON file
json_file_path = "user_data.json"
with open(json_file_path, "r") as json_file:
    user_data = json.load(json_file)

# Filter data by "CreditCardType" equals to "American Express"
filtered_data = [user for user in user_data if user.get("CreditCardType") == "American Express"]

# Specify the CSV file name
csv_file_path = "american_express_users.csv"

# Write the filtered data to a CSV file
with open(csv_file_path, "w", newline='') as csv_file:
    # Define the CSV writer
    csv_writer = csv.DictWriter(csv_file, fieldnames=filtered_data[0].keys())

    # Write the header
    csv_writer.writeheader()

    # Write the filtered data
    csv_writer.writerows(filtered_data)

print(f"Filtered data for 'American Express' users saved to {csv_file_path}")
```

6\. Load the file from this url: [https://www.dropbox.com/s/7u3lm737ogbqsg8/mushrooms_categorized.csv?dl=1](https://www.dropbox.com/s/7u3lm737ogbqsg8/mushrooms_categorized.csv?dl=1) with Pandas. 
+ Explore the data (see the info of the data)
+ Draw the istogram of the 'class' field. Decribe wath yuou see


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the URL
url = "https://www.dropbox.com/s/7u3lm737ogbqsg8/mushrooms_categorized.csv?dl=1"
mushrooms_data = pd.read_csv(url)

# Explore the data
print("Data Info:")
print(mushrooms_data.info())

# Draw the histogram of the 'class' field
plt.figure(figsize=(8, 6))
mushrooms_data['class'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Distribution of Mushroom Classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```
