2\. Find the outer product of the following two vecotrs

```python
u = np.array([1,3,5,7])
v = np.array([2,4,6,8])
```

Do this in the following ways:

   * Using the function outer in numpy
   * Using a nested for loop or list comprehension
   * Using numpy broadcasting operatoins


````python
import numpy as np

u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])

# Using np.outer function
outer_product_np = np.outer(u, v)

# Using a nested for loop or list comprehension
outer_product_loop = np.array([[u_i * v_j for v_j in v] for u_i in u])

# Using NumPy broadcasting operations
outer_product_broadcasting = u[:, np.newaxis] * v

# Displaying the results
print("Using np.outer function:\n", outer_product_np)
print("\nUsing nested for loop or list comprehension:\n", outer_product_loop)
print("\nUsing NumPy broadcasting operations:\n", outer_product_broadcasting)
````

5\. Create a matrix that shows the 10 by 10 multiplication table.

 * Find the trace of the matrix

 * Extract the diagnoal offset by 1 upwards (this should be ```array([ 2,  6, 12, 20, 30, 42, 56, 72, 90])```)

````python
import numpy as np

# Create a matrix that shows the 10 by 10 multiplication table
multiplication_table = np.arange(1, 11) * np.arange(1, 11)[:, np.newaxis]

# Find the trace of the matrix
trace = np.trace(multiplication_table)

# Extract the anti-diagonal
anti_diagonal = np.diagonal(np.flipud(multiplication_table))

# Extract the diagonal offset by 1 upwards
offset_diagonal = np.diagonal(multiplication_table, offset=1)

# Display the results
print("Multiplication Table:\n", multiplication_table)
print("\nTrace:", trace)
print("\nAnti-Diagonal:", anti_diagonal)
print("\nOffset Diagonal (upwards by 1):\n", offset_diagonal)
````

6\. Use broadcasting to create a grid of distances

Route 66 crosses the following cities in the US: Chicago, Springfield, Saint-Louis, Tulsa, Oklahoma City, Amarillo, Santa Fe, Albuquerque, Flagstaff, Los Angeles
The corresponding positions in miles are: 0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448

  * Construct a 2D grid of distances among each city along Route 66
  * Convert that in km (those savages...)

````python
import numpy as np

# Cities along Route 66
cities = np.array(['Chicago', 'Springfield', 'Saint-Louis', 'Tulsa', 'Oklahoma City', 'Amarillo', 'Santa Fe', 'Albuquerque', 'Flagstaff', 'Los Angeles'])

# Positions in miles
positions_miles = np.array([0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448])

# Construct a 2D grid of distances among each city along Route 66 using broadcasting
distances_miles = np.abs(positions_miles - positions_miles[:, np.newaxis])

# Conversion factor from miles to kilometers
miles_to_km = 1.60934

# Convert distances to kilometers
distances_km = distances_miles * miles_to_km

# Display the results
print("Cities along Route 66:\n", cities)
print("\n2D Grid of Distances (in miles):\n", distances_miles)
print("\n2D Grid of Distances (in kilometers):\n", distances_km)
````

7\. Prime numbers sieve: compute the prime numbers in the 0-N (N=99 to start with) range with a sieve (mask).
  * Constract a shape (100,) boolean array, the mask
  * Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
  * Apply the mask to obtain an array of ordered prime numbers
  * Check the performances (timeit); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

**N.B. the following exercises are meant to be solved only if you are familiar with the numpy random library. If not you can skip them (postponed for one of the next exercise sessions)**


````python
import numpy as np
import timeit

def sieve_of_eratosthenes(n):
    is_prime = np.ones(n + 1, dtype=bool)
    is_prime[0:2] = False  # 0 and 1 are not prime numbers

    for i in range(2, int(np.sqrt(n)) + 1):
        if is_prime[i]:
            is_prime[i*i:n+1:i] = False

    return np.nonzero(is_prime)[0]

# N = 99 to start with
N = 99

# Construct a shape (100,) boolean array, the mask
mask = np.ones(N + 1, dtype=bool)

# Identify the multiples of each number starting from 2 and set the corresponding mask element
for i in range(2, int(np.sqrt(N)) + 1):
    if mask[i]:
        mask[i*i:N+1:i] = False

# Apply the mask to obtain an array of ordered prime numbers
primes = np.nonzero(mask)[0]

# Check the performances using timeit
time_regular = timeit.timeit(lambda: np.nonzero(mask)[0], number=10000)
time_optimized = timeit.timeit(lambda: sieve_of_eratosthenes(N), number=10000)

# Display the results
print("Prime Numbers (Regular Sieve):\n", primes)
print("\nPerformance (Regular Sieve):", time_regular)
print("\nPerformance (Optimized Sieve of Eratosthenes):", time_optimized)
````

8\. Diffusion using random walk

Consider a simple random walk process: at each step in time, a walker jumps right or left (+1 or -1) with equal probability. The goal is to find the typical distance from the origin of a random walker after a given amount of time. 
To do that, let's simulate many walkers and create a 2D array with each walker as a raw and the actual time evolution as columns

  * Take 1000 walkers and let them walk for 200 steps
  * Use randint to create a 2D array of size walkers x steps with values -1 or 1
  * Build the actual walking distances for each walker (i.e. another 2D array "summing on each raw")
  * Take the square of that 2D array (elementwise)
  * Compute the mean of the squared distances at each step (i.e. the mean along the columns)
  * Plot the average distances (sqrt(distance\*\*2)) as a function of time (step)
  
Did you get what you expected?

````python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Number of walkers and steps
walkers = 1000
steps = 200

# Create a 2D array with values -1 or 1
random_steps = np.random.choice([-1, 1], size=(walkers, steps))

# Build the actual walking distances for each walker
distances = np.cumsum(random_steps, axis=1)

# Take the square of the distances (elementwise)
distances_squared = distances**2

# Compute the mean of the squared distances at each step
mean_distances_squared = np.mean(distances_squared, axis=0)

# Take the square root to get the average distances
average_distances = np.sqrt(mean_distances_squared)

# Plot the average distances as a function of time (step)
plt.plot(np.arange(1, steps + 1), average_distances)
plt.xlabel('Steps')
plt.ylabel('Average Distance')
plt.title('Random Walk - Average Distance Over Time')
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Download the data file
#!wget https://www.dropbox.com/s/3vigxoqayo389uc/populations.txt

# Check the content of the file
#!head populations.txt
#sto usando windows quindi !wget e !head non vanno

# Load the data into a 2D array
data = np.loadtxt('populations.txt')

# Create arrays for each column
year, hares, lynxes, carrots = data.T

# Plot the populations over the years
plt.figure(figsize=(10, 6))
plt.plot(year, hares, label='Hares', marker='o')
plt.plot(year, lynxes, label='Lynxes', marker='o')
plt.plot(year, carrots, label='Carrots', marker='o')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Populations of Hares, Lynxes, and Carrots Over the Years')
plt.legend()
plt.show()

# Compute the main statistical properties
mean_hares = np.mean(hares)
mean_lynxes = np.mean(lynxes)
mean_carrots = np.mean(carrots)

std_hares = np.std(hares)
std_lynxes = np.std(lynxes)
std_carrots = np.std(carrots)

correlation_hares_lynxes = np.corrcoef(hares, lynxes)[0, 1]
correlation_hares_carrots = np.corrcoef(hares, carrots)[0, 1]
correlation_lynxes_carrots = np.corrcoef(lynxes, carrots)[0, 1]

# Display the statistical properties
print(f'Mean Hares: {mean_hares}')
print(f'Mean Lynxes: {mean_lynxes}')
print(f'Mean Carrots: {mean_carrots}')
print(f'Std Hares: {std_hares}')
print(f'Std Lynxes: {std_lynxes}')
print(f'Std Carrots: {std_carrots}')
print(f'Correlation Hares-Lynxes: {correlation_hares_lynxes}')
print(f'Correlation Hares-Carrots: {correlation_hares_carrots}')
print(f'Correlation Lynxes-Carrots: {correlation_lynxes_carrots}')

# Which species has the highest population each year?
max_population_species = np.argmax(data[:, 1:], axis=1)
max_population_species_names = np.array(['Hares', 'Lynxes', 'Carrots'])
species_with_highest_population_each_year = max_population_species_names[max_population_species]

# Display the species with the highest population each year
print(f'Species with the highest population each year:\n{species_with_highest_population_each_year}')
````
